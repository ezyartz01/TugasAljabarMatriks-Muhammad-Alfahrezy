L2  Norm

1. buat vector x dengan ukuran 3x1

```python
import numpy as np
```

```python
x = np.array([25, 30, 35])
```

hitung nilai L2 norm dari vector x

```python
(25**2 + 30**2 + 35**2)**(1/2)
```

```python
np.linalg.norm(x)
```

2. buat vector v dengan elemen (3,4)

```python
v = np.array([3,4])
```

hitung nilai L2 norm dari vector v

```python
(3**2 + 4**2)**(1/2)
```

```python
np.linalg.norm(v)
```

tampilkan unit vector dari vector v

```python
norm_v = np.linalg.norm(v)
```

```python
unit_v = v / norm_v
unit_v
```

hitung panjang unit vector dari vector v

```python
np.linalg.norm(v)
```
